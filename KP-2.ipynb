```python
fig, ax = plt.subplots(nrows=4,ncols=4)
plt.tight_layout()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend()
```
